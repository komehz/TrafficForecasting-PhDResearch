```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12,10), dpi=300)

ticks = [0, 36, 72, 108, 144, 180, 216, 252, 288]
xtiks = [k+0.5 for k in range(0,35)]
xlbs = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']*5

a = sns.heatmap( 
            df1,
            xticklabels=False,
            cbar=False,
            cbar_kws={'fraction' : 0.01,
                    'label': 'flow (veh/5-Min)'}, # shrink colour bar
            cmap='viridis',
            robust=True,
            ax=ax[0],
            )

for i in ticks:
    ax[0].axhline(i, color='white', lw=0.7)
for j in range(0,df1.shape[1],7):
    ax[0].axvline(j, color='white', lw=0.7)
for j in range(0,df1.shape[1],1):
    ax[0].axvline(j, color='white', lw=0.3)
    

b = sns.heatmap(
            df2,
            xticklabels=False,
            cbar=False,
            cbar_kws={'fraction' : 0.01,
                    'label': 'flow (veh/5-Min)'}, # shrink colour bar
            cmap='viridis',
            robust=True,
            ax=ax[1],
            )
    
for i in ticks:
    ax[1].axhline(i, color='white', lw=0.7)
for j in range(0,df2.shape[1],7):
    ax[1].axvline(j, color='white', lw=0.7)
for j in range(0,df2.shape[1],1):
    ax[1].axvline(j, color='white', lw=0.3)


cmap = mpl.cm.viridis
norm = mpl.colors.Normalize(vmin=0, vmax=500)

cbar = fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap), ax = ax, label='flow (veh/5-Min)', shrink=0.6)

# ax[0].set_xticks(xtiks)
# ax[0].set_xticklabels(xlbs, rotation=90, va='center') #, rotation_mode='default')
ax[0].set_yticks([0, 36, 72, 108, 144, 180, 216, 252, 288]);
ax[0].set_yticklabels(['00:00', '03:00', '06:00', '09:00', '12:00', '15:00', '18:00', '21:00','00:00']);
ax[0].tick_params(which='both',labelsize= 15)
ax[0].set_ylabel("")
ax[0].set_xlabel("")

# Fixing the figure issue
import matplotlib.transforms
# Create offset transform by 5 points in x direction
dx = 0/72.; dy = -15/72. 
offset = matplotlib.transforms.ScaledTranslation(dx, dy, fig.dpi_scale_trans)

# apply offset transform to all x ticklabels.
for label in ax[0].xaxis.get_majorticklabels():
    label.set_transform(label.get_transform() + offset)


# ax[1].set_xticks(xtiks)
# ax[1].xaxis.tick_top()
ax[1].set_yticks([0, 36, 72, 108, 144, 180, 216, 252, 288]);
ax[1].set_yticklabels(['00:00', '03:00', '06:00', '09:00', '12:00', '15:00', '18:00', '21:00','00:00']);
ax[1].tick_params(which='both',labelsize= 15)
ax[1].set_ylabel("")
ax[1].set_xlabel("")
ax[1].figure.axes[-1].yaxis.label.set_size(15)
ax[1].figure.axes[-2].yaxis.label.set_size(15)
ax[1].figure.axes[-1].xaxis.label.set_size(15)
cbar.ax.tick_params(labelsize= 15);

#plt.tight_layout()
plt.plot()
plt.show()

# fig.savefig('Comparison2.png')
```
