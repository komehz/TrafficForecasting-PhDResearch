```python
import numpy as np
import math
import matplotlib.pyplot as plt
from colors import *
```

```python
x = np.arange(0,10_100,100)
```

```python
a = -2
b = 4*x + 25
c = (-25*x + 2*x**2)*-1

d = b**2-4*a*c # discriminant

lst1 = []
lst2 = []

for i in range(len(d)):
    if d[i] < 0:
        print ("This equation has no real solution")
    elif d[i] == 0:
        xa = (-b[i]+math.sqrt(b[i]**2-4*a*c[i]))/2*a
        print ("This equation has one solutions: "), xa
    else:
        x1 = (-b[i]+math.sqrt((b[i]**2)-(4*(a*c[i]))))/(2*a)
        x2 = (-b[i]-math.sqrt((b[i]**2)-(4*(a*c[i]))))/(2*a)
        lst1.append(x1)
        lst2.append(x2)
```

```python
y = np.asarray(lst1)
z = np.asarray(lst2)
time = x
```

```python
plt.rcParams.update({'font.size': 20})
```

```python
# Initialize figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Fill area when income <= expenses with red
ax.fill_between(
    time, y, z, where=(y <= z), 
    interpolate=True, color="green", alpha=0.25,
    label="GEH < 5"
)

# Plot lines
ax.plot(time, y, color=BLUE2, label="Lower limit", lw=3)
ax.plot(time, z, color=BLUE3, label="Upper limit", lw=3)

ax.legend(fontsize=20);
ax.grid();
ax.set_xlim([0, 10_000]);
ax.set_ylim([0, 10_000]);

# Set x-axis tick labels and rotation as integers
ticks = ax.get_xticks().astype(int)
ax.set_xticks(ticks)
ax.set_xticklabels(ticks, rotation=90, fontsize=20)

ax.set_ylabel('Predicted $(F_p)$', fontsize=20);
ax.set_xlabel('Actual $(F_a)$', fontsize=20);
```

```python
x[0] = 1

k = (z - x)/x;     # Relative error

x[0] = 0
k[0] = 100_000
```

```python
plt.figure(figsize=(8,8))
# Same, but add a stronger line on top (edge)
plt.fill_between(x, k, color="green", alpha=0.25, label="GEH < 5")
plt.plot(x, k, color=BLUE2, alpha=1, label="GEH = 5", lw=3)
# See the line plot function to learn how to customize the plt.plot function

plt.legend();
plt.grid();
plt.xlim([0, 10_000]);
plt.ylim([0, 0.6]);
plt.ylabel('Acceptable relative deviation (r.d.)', fontsize=20);
plt.xlabel('Actual $(F_a)$', fontsize=20);
plt.xticks(rotation=90)

# ax.set_title('Scores by group and gender');

# Show the graph
plt.show()
```
