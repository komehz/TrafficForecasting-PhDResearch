```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage,AnnotationBbox
from colors import *
```

```python
cities = ["London", "Bristol", "Manchester", "Birmingham", "Belfast", 
          "Nottingham", " Edinburgh", "Hull" ,"Leeds", "Leicester"]
valuesA = [1377, 805, 742, 646, 636, 625, 593, 604, 503, 551]
valuesA.reverse()
cities.reverse()
```

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=300)

ax.grid(linestyle=':', linewidth=1)
ax.barh(range(len(cities)), valuesA, height=0.6, align="center", color="salmon")
ax.set_yticks(range(len(cities)))
ax.set_yticklabels(cities)
ax.tick_params(axis='y', which='major', pad=5)
ax.set_xlim(0,1500)
ax.set_xlabel('Cost per Commuter')

ax.set_axisbelow(True)

    
def add_value_labels(ax, spacing=4):
    """Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """

    # For each bar: Place a label
    for i, rect in enumerate(ax.patches):
        # Get X and Y placement of label from rect.
        y_value = valuesA[i]
        x_value = rect.get_y()

        # Number of points between bar and label. Change to your liking.
        space = spacing
        # Vertical alignment for positive values
        va = 'baseline'


        # Use Y value as label and format number with one decimal place
        label =  ""+str(y_value) 
        
        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (y_value, x_value),         # Place label at end of the bar
            xytext=(15, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for
                                        # positive and negative values.


# Call the function above. All the magic happens there.
add_value_labels(ax)

plt.savefig('CongestionLevel-cost.png', bbox_inches='tight')
plt.show()
```
