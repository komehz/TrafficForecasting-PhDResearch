```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage,AnnotationBbox
from colors import *
```

```python
def get_flag(name):
    path = "country-flags-main\png100px\{}.png".format(name.title())
    im = plt.imread(path)
    return im
```

```python
def offset_image(coord, name, ax):
    img = get_flag(name)
    im = OffsetImage(img, zoom=0.3)
    im.image.axes = ax

    ab = AnnotationBbox(im, (coord, 0),  xybox=(0., -16.), frameon=False,
                        xycoords='data',  boxcoords="offset points", pad=0)

    ax.add_artist(ab)
```

```python
cities = ["Mumbai", "Jarkarta", "Dublin", "London", "Tokyo", "Tel-Aviv", "Shenzhen", "Kyiv" ,"Los-Angeles"]
valuesA = [65, 53, 48, 38, 42, 46, 37, 53, 42]
```

```python
fig, ax = plt.subplots(figsize=(8, 4), dpi=200)

ax.grid(linestyle=':', linewidth=1)
ax.bar(range(len(cities)), valuesA, width=0.5, align="center", color=BLUE4)
ax.set_xticks(range(len(cities)))
ax.set_xticklabels(cities)
ax.tick_params(axis='x', which='major', pad=26)
ax.set_ylim(0,70)
ax.set_ylabel('Congestion Level (%)')

ax.set_axisbelow(True)

countries = ["in", "id", "ie", "gb", "jp", "il", 
             "cn", "ua","us"]

for i, c in enumerate(countries):
    offset_image(i, c, ax)

plt.savefig('CongestionLevel.png', bbox_inches='tight')
plt.show()
```
