```python
import numpy as np
import matplotlib.pyplot as plt
from colors import *
```

```python
x_true = 100
x_pred = np.arange(50, 150, 0.05)
```

```python
y_mse = np.square(x_pred - x_true)
```

```python
y_mae = np.abs(x_pred - x_true)
```

```python
y_geh = np.abs(x_pred - x_true)/np.sqrt((x_pred + x_true)/2)
```

```python
plt.plot(x_pred, y_mse, "blue")
plt.grid(True, which="major")
plt.show()
```

```python
plt.plot(x_pred, y_mae, "blue")
plt.grid(True, which="major")
plt.show()
```

```python
plt.plot(x_pred, y_geh, "blue")
plt.grid(True, which="major")
plt.show()
```

```python
# MSE loss function
def mse_loss(y_pred, y_true):
    squared_error = (y_pred - y_true) ** 2
    sum_squared_error = np.sum(squared_error)
    loss = sum_squared_error / y_true.size
    return loss
    
# Plotting
x_vals = np.arange(-20, 20, 0.01)
y_vals = np.square(x_vals)

plt.plot(x_vals, y_vals, "blue")
plt.grid(True, which="major")
plt.show()
```

```python
# MAE loss function
def mae_loss(y_pred, y_true):
    abs_error = np.abs(y_pred - y_true)
    sum_abs_error = np.sum(abs_error)
    loss = sum_abs_error / y_true.size
    return loss
    
# Plotting
x_vals = np.arange(-100, 100, 0.01)
y_vals = np.abs(x_vals)

plt.plot(x_vals, y_vals, "red")
plt.grid(True, which="major")
plt.show()
```
