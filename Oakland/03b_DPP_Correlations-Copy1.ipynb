### (ii) Day pattern (mon - fri)

### Week 1

```python
df = TrainVal['f4'][(TrainVal.index.month == 8)]


df1 = df[df.index.day == 21].reset_index(drop=True)
df2 = df[df.index.day == 22].reset_index(drop=True)
df3 = df[df.index.day == 23].reset_index(drop=True)
df4 = df[df.index.day == 24].reset_index(drop=True)
df5 = df[df.index.day == 25].reset_index(drop=True)
#df6 = df[df.index.day == 26].reset_index(drop=True)
#df7 = df[df.index.day == 27].reset_index(drop=True)


wk1 = pd.DataFrame([df1,df2,df3,df4,df5]).T #,df6,df7]).T
wk1.columns = ['Mon','Tue','Wed','Thu','Fri'] #,'Sat','Sun']
```

```python
# The above as a plot

wk1 = wk1.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(wk1, dtype=bool), 1)

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 9))

# Draw the heatmap with the mask and correct aspect ratio
crr = sns.heatmap(wk1, mask=mask, cmap="rocket_r", fmt='.2g',
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

crr.set_yticklabels(crr.get_yticklabels(), rotation=0);
```
