# Correlations

### (i) Week pattern (mon - mon)

```python
df = TrainVal['f3'][(TrainVal.index.dayofweek == 0)][:12960]


df1 = df[df.index.day == 21].reset_index(drop=True)
df2 = df[df.index.day == 28].reset_index(drop=True)
df3 = df[df.index.day == 4].reset_index(drop=True)
df4 = df[df.index.day == 11].reset_index(drop=True)
df5 = df[df.index.day == 18].reset_index(drop=True)


mon = pd.DataFrame([df1,df2,df3,df4,df5]).T
mon.columns = [1,2,3,4,5]
```

```python
# The above as a plot

Mon = mon.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(Mon, dtype=bool), 1)

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 9))

# Draw the heatmap with the mask and correct aspect ratio
crr = sns.heatmap(Mon, mask=mask, cmap="rocket_r", fmt='.2g',
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

crr.set_yticklabels(crr.get_yticklabels(), rotation=0);
```
